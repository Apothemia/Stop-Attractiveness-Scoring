```python
import pandas as pd
import glob
from datetime import datetime
```

```python
search_dict = 'data/Bus systems/0307'
files = glob.glob(search_dict + '/*.csv')
```

```python
df = pd.concat(map(pd.read_csv, files))
df
```

```python
drop_cols = ['CITY_CODE', 'DRIVER_CODE', 'CHECKER_CODE', 'VAL_FLAG', 'RUN_COMP_CODE', 'LOW_COMP_CODE', 'GET_POINT_CODE', 'POS_CODE', 'UP_TIME']

df = df.drop(drop_cols, axis=1)
df
```

```python
datetime_cols = ['DEAL_TIME', 'CREATEDATE', 'UPDATEDATE']
date_format = '%Y%m%d%H%M%S'

for col in datetime_cols:
    df[col] = df[col].apply(lambda val: datetime.strptime(str(val), date_format))
df
```

```python
unique_passengers = df.groupby(by='BUSDATA_ID', group_keys=True)
len(unique_passengers)
```

```python
for key, item in unique_passengers:
    if len(item.index) > 2:
        display(item)
        break
```
